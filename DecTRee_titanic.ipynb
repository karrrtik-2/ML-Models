```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
target = df.Survived
```

```python
X = df[['Age','Fare','Sex','Pclass']]
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
X['gender'] = encoder.fit_transform(X['Sex'])
```

```python
X.head()
```

```python
X.drop('Sex',axis='columns',inplace=True)
```

```python
X['Age']=X.Age.fillna(X.Age.mean())
```

```python
X.Age.astype('int')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,target,train_size = 0.8,random_state=10)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
